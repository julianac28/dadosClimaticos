#### DADOS 2015 A 2018

```python
#Criar df com os dados do ano

dados_2015_2018 = pd.DataFrame()

for ano in range(2015,2019):
    
    # Lista de arquivos
    with ZipFile(str(ano)+".zip", "r") as lista:
        #print(*lista.namelist(), sep="\n")
        listaArquivos = lista.namelist()

    # Separar os arquivos de SC
    listaSC = []
    for i in range(0,len(listaArquivos)):
        if re.findall("INMET_S_SC", listaArquivos[i]):
            listaSC.append(listaArquivos[i])
            #print(listaArquivos[i])

    # Selecionar o arquivo na listaSC e indicar a linha do "header" ==> cabeçalho
    for i in listaSC:
        with ZipFile(str(ano)+".zip") as file:
            with file.open(i) as f:
                df = pd.read_csv(f, sep=";", header=8, encoding="ISO-8859-1", decimal=",")
    
            #Meta dados inicio do CSV, transpor para transformar em coluna
            with ZipFile(str(ano)+".zip") as file:
                with file.open(i) as f:
                    h = pd.read_csv(f, sep=";", nrows=7, encoding="ISO-8859-1",decimal=",").set_index("REGIÃO:").T
            
                    # Incluir os meta dados no final df
                    for i in range(0,len(h.columns)):
                        df[h.columns[i].replace(":","")] = h.iloc[0,i]
            
                    # Concatenas os arquivos
                    dados_2015_2018 = pd.concat([df, dados_2015_2018], ignore_index=True)
                    
#renomear colunas    
dados_2015_2018 = dados_2015_2018.rename(columns={'DATA (YYYY-MM-DD)':'Data','HORA (UTC)':'Hora UTC','RADIACAO GLOBAL (KJ/m²)':'RADIACAO GLOBAL (Kj/m²)','ESTAÇÃO':'ESTACAO', 'DATA DE FUNDAÇÃO (YYYY-MM-DD)':'DATA DE FUNDACAO'}) 
```

#### DADOS 2019 A 2023

```python
#Criar df com os dados do ano

dados_2019_2023 = pd.DataFrame()

for ano in range(2019,2024):
    
    # Lista de arquivos
    with ZipFile(str(ano)+".zip", "r") as lista:
        #print(*lista.namelist(), sep="\n")
        listaArquivos = lista.namelist()

    # Separar os arquivos de SC
    listaSC = []
    for i in range(0,len(listaArquivos)):
        if re.findall("INMET_S_SC", listaArquivos[i]):
            listaSC.append(listaArquivos[i])
            #print(listaArquivos[i])

    # Selecionar o arquivo na listaSC e indicar a linha do "header" ==> cabeçalho
    for i in listaSC:
        with ZipFile(str(ano)+".zip") as file:
            with file.open(i) as f:
                df = pd.read_csv(f, sep=";", header=8, encoding="ISO-8859-1", decimal=",")
    
            #Meta dados inicio do CSV, transpor para transformar em coluna
            with ZipFile(str(ano)+".zip") as file:
                with file.open(i) as f:
                    h = pd.read_csv(f, sep=";", nrows=7, encoding="ISO-8859-1", decimal=",").set_index("REGIAO:").T
            
                    # Incluir os meta dados no final df
                    for i in range(0,len(h.columns)):
                        df[h.columns[i].replace(":","")] = h.iloc[0,i]
            
                    # Concatenas os arquivos
                    dados_2019_2023 = pd.concat([df, dados_2019_2023], ignore_index=True)
                    
```

#### CRIAÇÃO tb_dados_climaticos

```python
# Concatenas os arquivos
df = pd.DataFrame()
df = pd.concat([dados_2015_2018,dados_2019_2023], ignore_index=True)
```

```python
# Listar as tabelas do Postgres

def tables_names(cur):
    cur.execute("SELECT table_name FROM information_schema.tables WHERE table_schema='public' ORDER BY table_name")
    table_names = cur.fetchall()
    return table_names
table_names= tables_names(cor)

print('Quantidade de tabelas: ', len(table_names))
for line in table_names:
    print(line)
```
